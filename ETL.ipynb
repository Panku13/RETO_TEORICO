```python
import pandas as pd
import os
import re
import unicodedata
```

```python
CARPETA_SALIDA = "salida"

if not os.path.exists(CARPETA_SALIDA):
    os.makedirs(CARPETA_SALIDA)
```

```python
df = pd.read_csv("netflix.csv", sep=';')
```

```python
# Columnas que vamos a limpiar
columnas = ["title", "director", "cast", "country"]

# Función para quitar tildes, Ñ y caracteres especiales
def limpiar_texto(texto):
    if pd.isna(texto):
        return texto

    # quitar tildes y convertir Ñ -> N
    texto = unicodedata.normalize("NFKD", str(texto))
    texto = texto.encode("ascii", "ignore").decode("utf-8")

    # dejar solo letras, numeros, espacios y algunos signos
    texto = re.sub(r"[^a-zA-Z0-9\s,.-]", "", texto)

    return texto


# Aplicar solo a esas columnas
for columna in columnas:
    df[columna] = df[columna].apply(limpiar_texto)


# Mostrar resultado
print(df[columnas].head(10))
```

Elimino Espacios

```python
for columna in df.select_dtypes(include="object").columns:
    df[columna] = df[columna].str.strip()
```

Elimino registros duplicados

```python
df = df.drop_duplicates()
```

Cambio valores nulos

```python
df["director"] = df["director"].fillna("Sin Director")
df["cast"] = df["cast"].fillna("Sin Actor")
df["country"] = df["country"].fillna("Sin País")
df["rating"] = df["rating"].fillna("Sin Clasificación")
df["date_added"] = df["date_added"].fillna("")
df["duration"] = df["duration"].fillna("")
df["listed_in"] = df["listed_in"].fillna("Sin Género")
```

Convertir fecha


```python
df["date_added"] = pd.to_datetime(df["date_added"],errors="coerce")

print(df.info())
print(df.head())
```

Separamos la duración

```python
def separar_duracion(valor):

    if pd.isna(valor):
        return 0, ""

    valor = str(valor)
    numero = re.findall(r"\d+", valor)
    unidad = re.findall(r"[A-Za-z]+", valor)

    if numero:
        numero = int(numero[0])
    else:
        numero = 0

    if unidad:
        unidad = " ".join(unidad)
    else:
        unidad = ""

    return numero, unidad


df[["duration_value", "duration_unit"]] = df["duration"].apply(
    lambda x: pd.Series(separar_duracion(x)))
```

CREO TABLA PAIS

```python
lista_paises = []

for fila in df["country"]:
    if pd.isna(fila):
        continue

    for pais in str(fila).split(","):
        pais = pais.strip()

        if pais != "":
            lista_paises.append(pais)

pais = pd.DataFrame(sorted(set(lista_paises)), columns=["country"])
#Elimino repetidos
pais.insert(0, "pais_id",range(1, len(pais)+1))

print(pais.head())
print(pais.tail())

#GUARDO CSV
pais.to_csv(os.path.join(CARPETA_SALIDA, "pais.csv"), index=False,
    encoding="utf-8-sig")
```

CREO TABLA PUENTES

```python
def crear_tabla_puente(
        df,
        pelicula,
        catalogo,
        columna_csv,
        columna_catalogo,
        id_catalogo,
        id_puente):

    relaciones = []

    for _, fila in df.iterrows():

        # Buscar la película
        peli_id = pelicula.loc[pelicula["show_id"] == fila["show_id"],"peli_id"].values[0]

        # Si el dato está vacío lo ignoramos
        if pd.isna(fila[columna_csv]):
            continue

        # Separar por comas
        for valor in str(fila[columna_csv]).split(","):
            valor = valor.strip()

            if valor == "":
                continue

            # Buscar el ID correspondiente
            catalogo_id = catalogo.loc[catalogo[columna_catalogo] == valor,id_catalogo].values[0]

            relaciones.append({id_catalogo: catalogo_id,"peli_id": peli_id})

    puente = pd.DataFrame(relaciones)

    puente.insert(0,id_puente,range(1, len(puente)+1))

    return puente
```

TABLA PEL_ACT

```python
pel_act = crear_tabla_puente(
    df=df,
    pelicula=pelicula,
    catalogo=actor,
    columna_csv="cast",
    columna_catalogo="nom_cast",
    id_catalogo="actor_id",
    id_puente="pelact_id"
)

#GUARDAR CSV
pel_act.to_csv(os.path.join(CARPETA_SALIDA, "pel_act.csv"),index=False,
    encoding="utf-8-sig"
)
```
